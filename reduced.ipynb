```python
import sympy as sp

def ReducedEchelonForm(A : sp.Matrix):

    m = A.rows # m行
    n = A.cols # n列

    M = A.copy()

    r = 0
    j = 0

    while r <  m and j < n:
        # r行以下でj列が0でない行pを探索
        p = next( (i for i in range(r,m) if M[i,j] != 0), None) 

        # pが存在しなければ（つまり，j列のr行以下が全て0なら）以下のif文は実行されず，次の列へ進む

        # M[p,j]を軸としてj列の他の成分を0にする
        if p is not None:

            # M[r,j]=0ならr行とp行を入れ替える
            if r != p: 
                M[r, : ], M[p, : ] = M[p, : ], M[r, : ]
            
            # r行をM[r,j]で割って，M[r,j]=1とする
            M[r, : ] *= 1/M[r,j] 

            # j列のr+1行以降を全て0にする
            for i in range(r+1,m): 
                M[i, : ] -= M[i,j]*M[r, : ] # i行にr行の -M[i,j] 倍を加える
            r += 1
        j += 1

    # ここまでで階段形になっている

    r -= 1

    # 各行の主成分の上を全て0にする
    while r > 0: # 主成分のある最下行から
        p = next(i for i in range(n) if M[r,i] != 0) # 主成分のある列pを探索
        for i in range(r):
            M[i, : ] -= M[i, p]*M[r, : ] # i行にp列の -M[i,p] 倍を加える
        r -= 1
    return M
```

```python
# 例
A = sp.Matrix([[1,1,1,-2,-1],[1,2,3,-4,-4],[3,0,-3,1,7],[0,-1,-2,-1,0]])
display(A)
ReducedEchelonForm(A)
```
